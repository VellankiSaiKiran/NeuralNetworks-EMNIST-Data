```python
import numpy as np
import pandas as pd
from google.colab import drive
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout,GaussianNoise, Activation
from tensorflow.keras import activations
from sklearn.model_selection import train_test_split
```

```python
trainpath = '/content/drive/My Drive/Data/MNIST-Data/emnist-letters-train.csv'
testpath = '/content/drive/My Drive/Data/MNIST-Data/emnist-letters-test.csv'
```

```python
train_df = pd.read_csv(trainpath)
test_df = pd.read_csv(testpath)
```

```python
y_train = train_df.iloc[:, 0]
X_train = train_df.iloc[:, 1:].values
y_test = test_df.iloc[:, 0]
X_test = test_df.iloc[:, 1:].values
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)
```

```python
np.shape(X_train)
```

```python
np.shape(y_train)
```

```python
np.shape(X_test)
```

```python
np.shape(y_test)
```

```python
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)
X_val = X_val.reshape(-1, 28, 28, 1)
```

```python
X_train = X_train.astype('float32') / 255
X_test = X_test.astype('float32') / 255
X_val = X_val.astype('float32') / 255
```

```python
y_train = y_train - 1
y_test = y_test - 1
y_val = y_val - 1

y_train = to_categorical(y_train, num_classes=26)
y_test = to_categorical(y_test, num_classes=26)
y_val = to_categorical(y_val, num_classes=26)
```

```python
np.shape(X_train)
```

```python
np.shape(y_train)
```

```python
model1 = Sequential([

    Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)),

    MaxPooling2D(pool_size=(2, 2)),

    Dropout(0.25),


    Conv2D(64, kernel_size=(3, 3), activation='relu'),

    MaxPooling2D(pool_size=(2, 2)),

    Dropout(0.25),


    Flatten(),

    Dense(128, activation='relu'),

    Dropout(0.5),

    Dense(26, activation='softmax')
])

model1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model1.summary()

```

```python
model2 = Sequential([


    GaussianNoise(0.1, input_shape=(28, 28, 1)),
    Conv2D(32, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.25),


    Conv2D(128, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.25),

    Flatten(),


    Dense(256, activation='tanh'),
    Dropout(0.5),


    Dense(26, activation='sigmoid')
])

model2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model2.summary()
```

```python
model3 = Sequential([

    Conv2D(32, kernel_size=(3, 3), activation='sigmoid', input_shape=(28, 28, 1)),

    MaxPooling2D(pool_size=(2, 2)),

    Dropout(0.25),


    Conv2D(64, kernel_size=(3, 3), activation='sigmoid'),

    MaxPooling2D(pool_size=(2, 2)),

    Dropout(0.25),


    Flatten(),

    Dense(128, activation='sigmoid'),

    Dropout(0.5),

    Dense(26, activation='softmax')
])

model3.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model3.summary()

```

```python
print(X_train.shape, y_train.shape, X_val.shape, y_val.shape)
```

```python
history = model1.fit(
    X_train, y_train,
    batch_size=128,
    epochs=10,
    verbose=1,
    validation_data=(X_val, y_val)
)
```

```python
score1 = model1.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score1[0])
print('Test accuracy:', score1[1])
```

```python
history = model2.fit(
    X_train, y_train,
    batch_size=128,
    epochs=10,
    verbose=1,
    validation_data=(X_val, y_val)
)
```

```python
score2 = model2.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score2[0])
print('Test accuracy:', score2[1])
```
